```python
import pandas as pd
import numpy as np
from ast import literal_eval
```

```python
books = pd.read_json('../data/metadata.json',lines=True)
books
```

```python
books = pd.read_csv('../data/books.csv', encoding="utf-8", usecols=range(1,10))
books
```

```python
tags = pd.read_json('../data/tags.json',lines=True)
tags
```

```python
tag_count = pd.read_json('../data/tag_count.json',lines=True)
tag_count
```

```python
table1 = pd.merge(books,tag_count,on='item_id',how="inner")
table1
```

```python
table2 = pd.merge(table1,tags,left_on='tag_id',right_on='id',how="inner").drop(columns=['id'])
table2
```

```python
books_tag = table2.groupby('item_id',dropna=True).agg(tags=('tag','unique'))
books_tag
```

```python
df_book = books.join(books_tag,on='item_id',how='inner')
df_book[df_book['item_id']==16827462]
```

```python
ratings = pd.read_json('../data/ratings.json',lines=True)
ratings
```

```python
user_books = pd.merge(ratings,df_book,on='item_id',how='inner')
user_books
```

```python
def get_user_tags(userId):
    user_df = user_books[user_books['user_id']==userId]
    tag_explode = user_df.explode('tags')
    tag_explode = tag_explode.groupby('tags').agg(appear=('tags','count'))
    tag_explode = tag_explode.sort_values(['appear'],ascending = False)
    return tag_explode

    
```

```python
get_user_tags(2)
```

```python
tag_df = get_user_tags(2)[:5]
result = pd.merge(table2,tag_df,left_on='tag',right_on=tag_df.index,how='inner')
result = result.groupby('item_id').agg(tags=('tag','unique')).reset_index()
result
```

```python
def calculate_rating(bookId):
    book_ratings = ratings[ratings['item_id']==bookId]['rating']
    return book_ratings.mean()
```

```python
calculate_rating(16827462)
```

```python
read_books = ratings[ratings['user_id']==2]['item_id']
result = result[~result['item_id'].isin(read_books)]
#result['average_rating']= result['item_id'].apply(lambda x: calculate_rating(x))
top = pd.merge(books, result,on='item_id',how='inner')
top = top.sort_values('average_rating',ascending=False)
top
```

```python
books["average_rating"] = books["item_id"].apply(lambda x: calculate_rating(x))
books
```

```python
books.to_csv("../data/books.csv", index=False)
```
